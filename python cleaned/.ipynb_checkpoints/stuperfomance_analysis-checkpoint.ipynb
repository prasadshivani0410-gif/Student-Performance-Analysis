## Data Cleaning

```python
# Replacing space between the column names by '_'
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('/', '_')
df.columns
```

## Feature Engineering

```python
df['avg_score']= df[['math_score', 'reading_score',
       'writing_score']].mean(axis=1)
```

```python
df.head()
```

```python
def performance_category(score):
    if score>=80:
        return 'High'
    elif score >= 50:
        return 'Medium'
    else:
        return 'Low'

df['performance_category'] = df['avg_score'].apply(performance_category)
    
```

```python
df.head()
```

```python
plt.figure(figsize=(4,4))
sns.barplot(x='gender',y='total_score',data=df,hue='gender', palette='coolwarm')
plt.xticks(rotation=0)
```

#### Female students scored higher marks then male student

## FACTOR ANALYSIS

### Q1 Does parental education level affect scores? 

```python
parent_education= df.groupby('parental_level_of_education')[['math_score', 'reading_score', 'writing_score', 'avg_score']].mean().sort_values('avg_score', ascending=False)
parent_education
```

```python
#  box plot (scores by parental education)
plt.figure(figsize=(12,5))
sns.barplot(x='parental_level_of_education',y='avg_score',data=parent_education, hue='parental_level_of_education',palette='coolwarm')
plt.title('Average Score by Parental Education Level')
plt.xlabel('Parental Education Level')
plt.ylabel('Average Score')
plt.xticks(rotation=0)
plt.show()
```

##### It shows a higher Education level have a better average score. This suggests that parental education may have a positive influence on student performance.

```python
# Q2 Do students who completed test preparation score higher?
test_prep = df.groupby('test_preparation_course')[['math_score', 'reading_score', 'writing_score', 'avg_score']].mean()
test_prep
```

```python
# bar chart (test prep comparison)
plt.figure(figsize=(4,4))
sns.barplot(x='test_preparation_course',y='avg_score',data=test_prep.reset_index(), hue='test_preparation_course', palette='coolwarm')
plt.title('Average Score by Test Preparation Course')
plt.xlabel('Test Preparation Course')
plt.ylabel('Average Score')
plt.xticks(rotation=0)
plt.show()
```

#### Those Students who completed the test preparation course achieved higher average scores. This shows that test preparation has a positive impact on performance. 

```python
# Q3 What is the correlation between reading, writing, and math scores?

correlation= df[['math_score', 'reading_score', 'writing_score']].corr()
correlation
```

```python
# correlation heatmap
sns.heatmap(correlation, annot=True, cmap="vlag")
```

#### Students who perform well in reading and writing have the strongest relationship.

```python
# Q4. Which gender performs better in which subject?
gender_analysis = df.groupby('gender')[['math_score', 'reading_score', 'writing_score']].mean()
gender_analysis
```

```python
# grouped bar (gender vs subject)
gender_analysis.plot(kind='bar', figsize=(8,5),colormap="Blues")
plt.title('Average Subject Scores by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Score')
plt.xticks(rotation=0)
plt.show()
```

#### Female Students are better in reading as well as writing subject and Male Students are better in maths subject.

```python
# Q5. What is the distribution of total scores?
#  histogram (total score distribution)
sns.histplot(df['total_score'], bins=50, kde=True)

plt.title('Distribution of Total Scores')
plt.xlabel('Total Score')
plt.ylabel('Number of Students')
plt.show()
```

```python
#  scatter plot (reading vs math)
sns.scatterplot(x='reading_score', y='math_score',data=df, hue='gender', palette='coolwarm')
plt.title('Reading Score vs Math Score')
plt.xlabel('Reading Score')
plt.ylabel('Math Score')
plt.show()
```

```python
# box plot (scores by parental education)
plt.figure(figsize=(12,5))
sns.boxplot(x='parental_level_of_education', y='total_score',data=df,hue='parental_level_of_education' ,palette='coolwarm')
plt.title('Total Score Distribution by Parental Education Level')
plt.xlabel('Parental Education Level')
plt.ylabel('Total Score')
plt.xticks(rotation=0)
plt.show()
```

## AT-RISK STUDENT SEGMENTATION

```python
df['at_risk'] = np.where(
    (df['math_score'] < 50) | 
    (df['reading_score'] < 50) | 
    (df['writing_score'] < 50),
    'Improve',
    'Good'
)

df.head()
```

```python
df['at_risk'].value_counts()
```

```python
df['at_risk'].value_counts(normalize=True)*100
```

#### 18.8% students needed extra academic support.

```python
df
```

```python
# Risk due to lunch
risk_bylunch=df.groupby('lunch')['at_risk'].value_counts(normalize=True).unstack()*100
risk_bylunch
```

```python
plt.figure(figsize=(4,4))
sns.barplot(x='lunch', y='Improve',data = risk_bylunch, hue='lunch',palette='coolwarm')
plt.title('Risk percentage by Lunch Type')
plt.xlabel('Lunch Type')
plt.ylabel('At-Risk')
```

#### Free/reduced lunch students need to improve more as compared to students having standard lunch.

```python
# Risk due to test_preparation_course
risk_byprep=df.groupby('test_preparation_course')['at_risk'].value_counts(normalize=True).unstack()*100
risk_byprep
```

```python
plt.figure(figsize=(4,4))
sns.barplot(x='test_preparation_course', y='Improve',data = risk_byprep, hue='test_preparation_course', palette='coolwarm')
plt.title('Risk percentage by Test preparation Course')
plt.xlabel('Test preparation')
plt.ylabel('At-Risk')
```

#### Students who did not complete the test preparation need to improve more. Suggestions to provide test preparation support.

## PRINCIPAL'S REPORT
#### Summary
This project analysis was used to understand the students' performance and identify the factors affecting their academic results. This dataset contains 1000 records with demographic data and scores in maths, reading and writing. In this analysis parantal education, lunch, test preparation are the factors affecting their performance.

#### Key Findings
1. Higher education level (Master's degree) has a better average score. This suggests that parental education may have a positive influence on student performance.
2. Those Students who completed the test preparation course achieved higher average scores. This shows that test preparation has a positive impact on performance. 
3. Female Students have a higher total score, and they are better in reading and writing subjects. Male Students are slightly better in maths subjects.
4. A maximum number of students have a score close to 200.
5. Free/reduced lunch students need to improve more compared to students having standard lunch.
6. Reading and writing subjects have a strong correlation of 0.95.

####  3 Specific Actionable recommendations the school can implement.
1. The school should provide a structured test preparation, like mock tests for all students, especially for the students who have not completed any preparation course.
2. The school should provide additional academic, scholarship and counselling support for students from weaker socio-economic backgrounds, especially those with free/reduced lunch.
3. At-risk students who score below 50 in any subject should provide mentorship, doubt solving sessions, and regular progress tracking.
